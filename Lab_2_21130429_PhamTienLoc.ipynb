#2. Load dataset

```python
# code
dataset = pd.read_csv("student-dataset.csv")
dataset.head(10)
```

#4. Sort dataset by ***nationality***

#6. Use scatter plot to represent the relationship between ***age*** and ***refletter.rating***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
# code
plt.scatter(dataset['age'],dataset['refletter.rating'], color="green")
# x-axis label
plt.xlabel('age')
# frequency label
plt.ylabel('refletter.rating')
# plot title
plt.title('My scatter plot!')
plt.show()
```

#7. Use scatter plot to represent the relationship between ***age*** and ***sciences.grade***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
#code
plt.scatter(dataset['age'],dataset['sciences.grade'], color="red")
# x-axis label
plt.xlabel('age')
# frequency label
plt.ylabel('sciences.grade')
# plot title
plt.title('My scatter plot!')
plt.show()
```

#8. Use histogram plot to represent the distribution of ***Age***
using bins=5

```python
#code
# setting the ranges and no. of intervals
range = (0, 100)
bins = 5

# plotting a histogram
plt.hist(dataset['age'], bins, range, color='green',histtype='bar',rwidth=0.8)
plt.show()
```

#9. Create a data frame that computes the average of ages in each countries?

Hint: Use groupby, select the age column and aggregate using mean.

```python
#code
dataset_avg_age_by_country = dataset.groupby('nationality')['age'].mean().reset_index(name='average_age')
dataset_avg_age_by_country
```

#10. Create a dataframe including average of ***english.grade, math.grade, sciences.grade, language.grade, portfolio.rating, coverletter.rating, refletter.rating*** of each country


```python
#code
columns_to_average = ['english.grade', 'math.grade', 'sciences.grade', 'language.grade', 'portfolio.rating', 'coverletter.rating', 'refletter.rating']
dataset_avg_by_country = dataset.groupby('nationality')[columns_to_average].mean().reset_index()
dataset_avg_by_country
```
